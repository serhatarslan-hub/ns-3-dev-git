```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 8.0)
```

```python
cwnd_times = []
cwnds = []
with open('TcpVariantsComparison-cwnd.data','r') as f:
    for line in f:
        time, cwnd = line.split()
        cwnd_times.append(float(time))
        cwnds.append(float(cwnd))
```

```python
plt.plot(cwnd_times,cwnds)
plt.show()
```

```python
rtt_times = []
rtts = []
with open('TcpVariantsComparison-rtt.data','r') as f:
    for line in f:
        time, rtt = line.split()
        rtt_times.append(float(time))
        rtts.append(float(rtt))
```

```python
plt.plot(rtt_times,rtts)
plt.show()
```
